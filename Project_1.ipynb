```python
#open text file and assign to variable "myfile"
myfile = open("leavesofgrass.txt", 'r')
#read file
text = myfile.read()
```

```python
#assign open file to variable "document"
document = open('leavesofgrass.txt', 'r')
#create dictionary 
d = dict()
#for loop to iterate through dictionary
for line in document:
    #strip text, lower function to convert text, split text into individual tokens
    line = line.strip()
    line = line.lower()
    words = line.split(" ")
    #for loop to count words in dictionary 
    for word in words:
        if word in d:
            d[word] = d[word] +1
        else:
            d[word] = 1
#for all the key value pairs, print keys which are number of occurances of a word in the dictionary
for key in list(d.keys()):
    print(key, ':', d[key])
```

```python
#counting the words, sorted in descending order by order or occurance. Print statement to demonstrate Zipf's Law of Abbreviation
word_count = ((value,key) for (key,value) in d.items())
frequency = sorted(word_count, reverse=True)
print(frequency)
```
