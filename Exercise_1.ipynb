```python
# open text file and assign to variable 'document'
document = open('leavesofgrass.txt', 'r')
#create dictionary
d = dict()
#for loop to iterate over text, to strip whitespace, to make text all lowercase and to split into tokens
for line in document:
    line = line.strip()
    line = line.lower()
    words = line.split(" ")
# counts words in document
    for word in words:
        if word in d:
            d[word] = d[word] +1
        else:
            d[word] = 1
# outputs the word count in descending order via their frequency in the corpus
word_count = ((value,key) for (key,value) in d.items())
frequency = sorted(word_count, reverse=True)
print(frequency)
```
